## 1.4  task-5 特征工程
本次收获最大得小节，之前对数据挖掘的理解过于依赖算法，而忽视了特征的可解释性；整个挖掘过程从最基础的探索性分析，为提取特征提供了思路（是否周末等）。另外多重共线、不独立等在统计学建模看重的问题在机器学习中建模的应用。
